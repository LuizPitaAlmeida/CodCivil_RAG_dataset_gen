```python
import pandas as pd
import os
from typing import List

results = [file for file in os.listdir('.') if (file.endswith('.csv')) & (file[:6] == 'df_rag')]
results = {file.split('.')[0]: pd.read_csv(file) for file in results}
results['df_rag_art_inventado_full'].columns
```

```python
def get_df(columns: List[str]=[None]):
    df_results = []
    for name, df in results.items():
        if columns[0] is not None:
            df = df[columns]
        df.columns = [name + '_' + col for col in df.columns if col in columns]
        df_results.append(df)

    df_results = pd.concat(df_results, axis=1)
    return df_results
```

```python
artid_df = get_df(['artigos_id'])
artid_df
```

```python
def convert_strlist_to_liststr(strlist):
    strlist = strlist[1:-1] # remove []
    strlist = strlist.replace("'", "").replace(".", "").replace(" do Código Civil", "").replace(" do CC", "").replace(" CC", "").replace("/2002", "").replace("/02", "").strip() # remove '
    liststr = strlist.split(", ") # split elements
    return liststr
```

```python
join = artid_df.applymap(convert_strlist_to_liststr).sum(axis=1)
join
```

```python
unique = join.apply(lambda x: list(set(x)))
pd.DataFrame(unique, columns=['unique'])
```

```python
from collections import Counter

counts = []
selected = []
for idx, row in join.items():
    count = Counter(row)
    counts.append(count)
    selected.append([element for element, c in count.items() if c > 2]
)

counts
```
