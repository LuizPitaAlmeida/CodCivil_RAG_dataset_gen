````python
from crewai import Agent

question_editor = Agent(
    llm=llama3,
    role='Dataset Formatter',
    goal='Get a {question}, split its main parts and organize it in JSON output format.',
    verbose=True,
    memory=False,
    backstory=(
    "Lets create a dataset of Question Answers Task using contest questions."
    "The idea is convert each topic in a Wrong - Right question keeping the context"
    ),
    tools=[],
    allow_delegation=False
)


from crewai import Task

# Research task
full_question_edit_task = Task(
  description=(
    "Split the {question} into Header, Questions ans Answer."
    "The Header normally ends in the sentence starting with 'Prova:'"
    "The questions could be multiple choice alternatives or 'Certo e Errado'"
    "In the case of multiple choices each alternive will generate a new question"
    "by repeating the context, the alternative, and asking if it is Right or Wrong"
    "The answer could be a letter or topics indication like I and IV, your job is"
    "generate a 'Right or Wrong' answer for each alternative in case of multiplechoice"
    "matching as Right the alternatives pointed in the answer field of the input"
    "sinalized after 'OpÃ§Ã£o correta:'. You are not allowed to change the answer given."
    "Example 1:"
    "1\nQ##### LAW_NAME\nAno: YEAR\nBanca: BANCA_NAME\nProva: PROVA_NAME"
    "Contexto [...] Diante deste fato, assinale a afirmativa correta."
    "A) [...] B) [...] C) [...] D) [...] E) [...]"
    "Responder\nErrada! OpÃ§Ã£o correta: A"
    "header = Q#####, LAW_NAME, Ano: YEAR, Banca: BANCA_NAME, Prova: PROVA_NAME"
    "questions = [context: CONTEXT, question: Text_of_A, [...], context: CONTEXT, question: Text_of_E]"
    "answers = [right, [...], wrong]"
  ),
  expected_output='A JSON format with fields: "header": header, "questions": [list_of_questions], "answers":[list_of_answers]',
  tools=None,
  agent=question_editor,
)

from crewai import Crew, Process

# Forming the tech-focused crew with some enhanced configurations
crew = Crew(
  agents=[question_editor],
  tasks=[full_question_edit_task],
  process=Process.sequential,  # Optional: Sequential task execution is default
  memory=False, # memory default True needs openai
  cache=True,
  max_rpm=80,
  share_crew=True,
  verbose=2
)
````

````python
from crewai_tools import tool

@tool("Multiple-Choice Question Splitter")
def mc_question_splitter(question: str) -> str:
    """Split a Multiple-Choice Question into HEADER, CONTEXT, QUESTIONS, ANSWERS."""
    prompt = (
        "Split a Multiple-Choice Question into HEADER, CONTEXT, QUESTIONS, ANSWERS."
        "The Header normally ends in the sentence starting with 'Prova:'"
        "The questions are multiple choice alternatives, so each alternive will generate a new question"
        "by repeating the context, the alternative, and adding it is 'correto' or 'errado' at the end."
        "The answer could be a letter or topics indication like I and IV, your job is"
        "generate a 'Right or Wrong' answer for each alternative"
        "matching as Correto the alternatives pointed in the answer field of the input"
        "sinalized after 'OpÃ§Ã£o correta:'. You are not allowed to change the answer given."
    )
    prompt = PromptTemplate(template=prompt + "\n{format_instructions}\n{question}\n", input_variables=["question"], partial_variables={"format_instructions": json_parser.get_format_instructions()})
    chain = prompt | llama3_70 | json_parser
    return chain.invoke({"question": question})

@tool("Binary Right-Wrong Question Splitter")
def bin_rw_question_splitter(question: str) -> str:
    """Split a Binary Right-Wrong Question into HEADER, CONTEXT, QUESTION, ANSWER."""
    prompt = (
        "Split a Binary Right-Wrong Question into HEADER, CONTEXT, QUESTION, ANSWER."
        "The Header normally ends in the sentence starting with 'Prova:'"
        "The original question are composed by CONTEXT + Two alternatives ('Certo', 'Errado') so the new question"
        "will be formed by repeating the context, and adding 'Correto ou Errado?' at the end."
        "The answer sinalized after 'OpÃ§Ã£o correta:' will be the letter that indicates if it is 'correto' or 'errado'"
        "your job is write 'correto' or 'errado' in answer field accordingly to that letter."
        "Notice that even having two alternatives only one question will be generated."
        "You are not allowed to change the answer given."
    )
    prompt = PromptTemplate(template=prompt + "\n{format_instructions}\n{question}\n", input_variables=["question"], partial_variables={"format_instructions": json_parser.get_format_instructions()})
    chain = prompt | llama3_70 | json_parser
    return chain.invoke({"question": question})

@tool("True-or-False Assessments Question Splitter")
def tf_ass_question_splitter(question: str) -> str:
    """Split a True-or-False Assessments Question into HEADER, CONTEXT, QUESTIONS, ANSWERS."""
    prompt = (
        "Split a True-or-False Assessments Question into HEADER, CONTEXT, QUESTIONS, ANSWERS."
        "The Header normally ends in the sentence starting with 'Prova:'"
        "The CONTEXT will be everything before the list of assertivas (assertions)."
        "Each assertiva will form a separate QUESTION, which is the assertiva itself."
        "The ANSWER field will be either 'V' for verdadeiro or 'F' for falso."
        "Your job is to mark each assertion as 'V' or 'F' based on the provided correct sequence."
        "The sequence of correct answers will be indicated after 'OpÃ§Ã£o correta:'."
        "Match the assertions to the provided correct sequence without changing the original answers."
    )
    prompt = PromptTemplate(template=prompt + "\n{format_instructions}\n{question}\n", input_variables=["question"], partial_variables={"format_instructions": json_parser.get_format_instructions()})
    chain = prompt | llama3_70 | json_parser
    return chain.invoke({"question": question})

@tool("Statement Question Splitter")
def stats_question_splitter(question: str) -> str:
    """Split a Statement Question into HEADER, CONTEXT, QUESTIONS, ANSWERS."""
    prompt = (
        "Split the question into HEADER, CONTEXT, QUESTIONS, ANSWERS."
        "The HEADER normally ends in the sentence starting with 'Prova:'"
        "The CONTEXT includes everything before the list of afirmativas (statements)."
        "Each afirmativa will form a separate QUESTION, which is the the context, the afirmativa itself and a sentence 'Correto ou Errado?'."
        "The ANSWER field will indicate whether the afirmativa is included in the correct response or not."
        "Your job is to mark each afirmativa as 'Correto' or 'Errado' based on the correct alternative provided."
        "The correct answer is indicated after 'OpÃ§Ã£o correta:'"
        "Match the afirmativas to the provided correct alternative without changing the original answer."
    )
    prompt = PromptTemplate(template=prompt + "\n{format_instructions}\n{question}\n", input_variables=["question"], partial_variables={"format_instructions": json_parser.get_format_instructions()})
    chain = prompt | llama3_70 | json_parser
    return chain.invoke({"question": question})
````
